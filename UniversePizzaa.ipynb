```python
import csv
import datetime

print("Universe Pizza'ya Hoşgeldiniz!<\n")


#menu.txt oluşturma 
with open('Menu.txt', 'w') as file: 
  file.write('Pizza Çeşitleri:\n 1: Klasik Pizza\n 2: Margarita Pizza\n 3: Türk Pizza\n 4: Sade Pizza\n')
  file.write('Sos Çeşitleri:\n 11: Zeytin\n 12: Mantar\n 13: Keçi Peyniri\n 14: Et\n 15: Soğan\n 16: Mısır\n')
  file.write('Teşekkür Ederiz!')

menu = open('Menu.txt')
print(menu.read()) # oluşturulan menu.txt dosyasının tamamını output olarak veriyor

#üst Sınıf
class Pizza:
  def get_description(self):
    return self.__class__.__name__

  def get_cost(self):
    return self.__class__.cost


# Pizza'nın alt sınıfları
class KlasikPizza(Pizza):
  
  cost = 20.5
  
  def __init__(self):
    self.description = "Sucuk, Salam, Biber, Zeytin, Mantar, Kaşar Peyniri"

class MargaritaPizza(Pizza):
  
  cost = 18

  def __init__(self):
    self.description = "Domates, Mozzarella, Fesleğen, Kaşar Peyniri"

class TurkPizza(Pizza):
  
  cost = 25.5
  
  def __init__(self):
    self.description = "Et, Biber, Patlıcan, Beyaz Peynir, Kaşar Peyniri"

class SadePizza(Pizza):

  cost = 16
  
  def __init__(self):
    self.description = "Domates, Sucuk, Kaşar Peyniri"

# Süsleme/Sos sınıfları
class Decorator(Pizza):
    
  def __init(self,sos): 
    self.component = sos

  def get_cost(self):
    return self.component.get_cost() + Pizza.get_cost(self)

  def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)

class Zeytin(Decorator):

  cost = 2
  
  def __init__(self, sos):
    Decorator.__init__(self)
    self.component = sos

class Mantar(Decorator):
  
  cost = 1

  def __init__(self, sos):
    Decorator.__init__(self)
    self.component = sos 

class KeciPeyniri(Decorator):
  
  cost = 3

  def __init__(self,sos):
    Decorator.__init__(self)
    self.component = sos 

class Et(Decorator):
  
  cost = 4

  def __init__(self,sos):
    Decorator.__init__(self)
    self.component = sos   

class Sogan(Decorator):
  
  cost = 2

  def __init__(self,sos):
    Decorator.__init__(self)
    self.component = sos 

class Misir(Decorator):
  
  cost = 1.5

  def __init__(self,sos):
    Decorator.__init__(self)
    self.component = sos   

def main():
  
  # piiza seçimi
  while True:
    pizza_cesitleri = {1: KlasikPizza, 2: MargaritaPizza, 3: TurkPizza, 4: SadePizza }
    secim = input("Lütfen bir pizza seçiniz: ")
    if secim in ["1", "2", "3", "4"]:
      siparis = pizza_cesitleri[int(secim)]()
      break
    else:
      print("Geçersiz Seçim!")
      
  # sos seçimi
  while True:
    sos_cesitleri = {11: Zeytin, 12: Mantar, 13: KeciPeyniri, 14: Et, 15: Sogan, 16: Misir}
    secim = input("Lütfen bir sos seçiniz (Sos seçiminden çıkmak için 0'a basınız!!): ")
    if secim in ["11","12","13","14","15","16"]:
      siparis = sos_cesitleri[int(secim)](siparis)
      continue
    elif secim == "0":
      break
    else:
      print("Geçersiz Seçim! Tekrar deneyiniz.")

  # sipariş yazdırma 
  print("\n" + siparis.get_description().strip() + ":  " + str(siparis.get_cost()) + " TL ")

  # kullanıcıdan bilgi alma 
  isim = input("İsminizi giriniz: ")
  TC = input("TC kimlik numaranızı giriniz: ")
  kk_num = input("Kredi kartı numaranızı giriniz: ")
  kk_cvc = input("3 rakamlı cvc kodunu giriniz: ")
  siparis_zamani = datetime.datetime.now()

  # kullanıcıdan alınan bilgileri database'e aktarma 
  with open('Orders_Database.csv', 'a') as siparisler:  # 'a' -> Append only
    siparisler = csv.writer(siparisler, delimiter=',')
    siparisler.writerow([isim, TC, kk_num, kk_cvc, siparis.get_description(), siparis_zamani])

  # en son alınan sipariş detaylarını ekrana yazdırma
  with open('Orders_Database.csv', 'r') as siparisler:
    son_siparis = siparisler.readlines()[-1]
    print(son_siparis)

  # with open('Orders_Database.csv', 'w') as siparisler:
   # siparisler.write('')

if __name__ == '__main__':
  main()
```
